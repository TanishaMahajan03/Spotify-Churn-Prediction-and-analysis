# ***DATASET CLEANING***

**Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Import Dataset**

```python
df = pd.read_csv("spotify_churn_dataset.csv")
df.sample(5)
```

**Check Null Values**

```python
df.isnull().sum()
```

```python
df = df[df['avg_daily_minutes'] >= 0]
```

```python
df.duplicated().sum()
```

```python
df['subscription_type'] = df['subscription_type'].str.strip().str.lower()
df['top_genre'] =df['top_genre'].str.strip().str.lower()
df['country'] = df['country'].str.strip().str.upper()
```

```python
df.describe()
```

```python
df.sample(5)
```

**Encode Categorical Variables**

```python
df= pd.get_dummies(df, columns=['subscription_type', 'top_genre', 'country'], drop_first=True)
```

```python
df.sample(5)
```

```python
df['churned'].value_counts()
```

**Save Cleaned DATASET**

```python
df.to_csv("spotify_cleaned.csv", index=False)
```
